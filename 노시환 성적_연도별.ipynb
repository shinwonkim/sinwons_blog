```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Dash 앱 생성
app = dash.Dash(__name__)

# 드롭다운 메뉴 생성
dropdown_options = [{'label': col, 'value': col} for col in 노시환_성적_연도별_index.columns[1:]]
dropdown_default = [노시환_성적_연도별_index.columns[1]]

# 레이아웃 구성
app.layout = html.Div([
    html.H1("연도별 선수 성적 변화"),
    dcc.Dropdown(
        id='column-dropdown',
        options=dropdown_options,
        multi=True,
        value=dropdown_default
    ),
    dcc.Graph(id='line-plot')
])

# 콜백 함수 정의
@app.callback(
    Output('line-plot', 'figure'),
    [Input('column-dropdown', 'value')]
)
def update_graph(selected_columns):
    fig = px.line(노시환_성적_연도별_index, x='연도', y=selected_columns, title='연도별 선수 성적 변화')
    return fig

# 앱 실행
if __name__ == '__main__':
    app.run_server(debug=True)
```
